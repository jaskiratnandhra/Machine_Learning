# Text Summarizer Using Sentiment Analysis

```python
from langdetect import detect
from textblob import TextBlob
import nltk
from nltk.tokenize import sent_tokenize 
from googletrans import Translator
translator= Translator()

class code(object):
    def __init__(self):
        self.text=''
        self.l=[]
        self.ol=[]
        self.senti=[]
        self.final=''
        self.f=0
        self.get_input()
    def get_input(self):
        self.text=input("Enter: ")
        if(detect(self.text)=='en'):
            pass
        elif(detect(self.text)=='ta'):
            self.f=1
            self.ol=sent_tokenize(self.text)
            s=''
            for i in self.ol:
                
                tra = translator.translate(i)
                s+=tra.text+" "
            self.text=s
            
            
            
        
        self.preprocessing()
    def preprocessing(self):
        self.l=sent_tokenize(self.text)

        self.sent()
    def sent(self):
        for i in range(len(self.l)):
            obj=TextBlob(self.l[i])
            sentiment=obj.sentiment.polarity
            self.senti+=[sentiment]
        
        self.display()
    def display(self):
        if(self.f==0):
            for i in range(len(self.senti)):
                if(self.senti[i]>0.0):
                    self.final+=self.l[i]
        else:
            for i in range(len(self.senti)):
                if(self.senti[i]>0.0):
                    self.final+=self.ol[i]
        print("\n"*3)
        print("The Summarizer text is :\n")
        print(self.final)
            






a=code()
    
    
    


    
```
